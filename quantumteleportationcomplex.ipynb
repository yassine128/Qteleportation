```python
from qiskit import *
```

```python
circuit = QuantumCircuit(5, 5)
```

```python
%matplotlib inline
circuit.draw(output='mpl')
```

```python
circuit.x(0)
#circuit.x(1)
```

```python
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.h(2)
circuit.cx(2,3)
circuit.cx(0,2)
circuit.h(0)
circuit.barrier()
circuit.reset(2)
circuit.h(2)
circuit.cx(2,4)
circuit.cx(1,2)
circuit.h(1)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure(0,0)
circuit.measure(1,1)
circuit.draw(output='mpl')
```

```python
circuit.measure(2,2)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.cx(2,3)
circuit.cx(2,4)
circuit.cz(0,3)
circuit.cz(1,4)
circuit.draw(output='mpl')
```

```python
circuit.measure(3,3)
circuit.measure(4,4)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
